```python
from selenium import webdriver
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://queapetito.com/ensaladas/'
ensaladas = webdriver.Firefox(executable_path='/home/dsc/Selenium/geckodriver')
ensaladas.get(url)
```

```python
#realizo el webscrapping de la página de recetas, sección ensaladas
```

```python
ensaladas.page_source
```

```python
ensaladas_html = ensaladas.page_source

len(ensaladas_html) 
```

```python
salad = BeautifulSoup(ensaladas_html, 'html5lib')
name_salad = salad.find_all('a', {'class':'title'}) #busco la clase titulo que es donde está la informacion de la url y el nombre
```

```python
ensaladas_link = list(name_salad)
```

```python
link = []
name_food = []
for i in ensaladas_link:
    for x in str(i).split('"'):
        if x[0:4]== 'http':
            link.append(x)
        if x[0]== '>':
            name_food.append(x[1:-4])
```

```python
name_food #verifico que la lista es correcta
```

```python
link #verifico que la lista es correcta
```

```python
r = requests.get(link[0])
salad = BeautifulSoup(r.content, 'html5lib')
salad.find_all('div', {'class':'comun-data nutrition'}) #informacion donde se encuentran las calorias
```
